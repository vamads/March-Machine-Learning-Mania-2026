# March Machine Learning Mania 2026 — Data Exploration

This notebook loads every CSV from the competition dataset into a pandas DataFrame so you can inspect schemas, shapes, and sample rows interactively.

```python
import pandas as pd
import os
from pathlib import Path

DATA_DIR = Path('march-machine-learning-mania-2026')
print(f'Data directory: {DATA_DIR.resolve()}')
print(f'Files found: {len(list(DATA_DIR.glob("*.csv")))}')
```

## Load All Datasets

Each CSV is loaded into a dictionary keyed by filename (without `.csv`). A summary table shows the shape and column list for every dataset.

```python
datasets = {}
for f in sorted(DATA_DIR.glob('*.csv')):
    name = f.stem  # filename without extension
    datasets[name] = pd.read_csv(f)

# Summary table
summary = pd.DataFrame([
    {'Dataset': name, 'Rows': df.shape[0], 'Cols': df.shape[1], 'Columns': ', '.join(df.columns)}
    for name, df in datasets.items()
])
summary
```
